```python
existe_gallinero = df_copy["Existegallinero"]
existe_gallinero.describe()
```

```python
print("Looking for outliers: {}".format(existe_gallinero.isnull().sum()))
```

```python
gallinero0_infestado = df_copy.query("Existegallinero == 0 & ((INFESTADO == 1) | (INFESTADO == 2) | (INFESTADO == 3))").shape[0]
gallinero1_infestado = df_copy.query("Existegallinero == 1 & ((INFESTADO == 1) | (INFESTADO == 2) | (INFESTADO == 3))").shape[0]
gallinero2_infestado = df_copy.query("Existegallinero == 2 & ((INFESTADO == 1) | (INFESTADO == 2) | (INFESTADO == 3))").shape[0]

gallinero0_no_infestado = df_copy.query("Existegallinero == 0 & INFESTADO == 0").shape[0]
gallinero1_no_infestado = df_copy.query("Existegallinero == 1 & INFESTADO == 0").shape[0]
gallinero2_no_infestado = df_copy.query("Existegallinero == 2 & INFESTADO == 0").shape[0]
```

```python
df_bar_plot_pclass_0 = pd.DataFrame({"Existegallinero": [0 ,1, 2], "frequency": [gallinero0_infestado+gallinero0_no_infestado, gallinero1_infestado+gallinero1_no_infestado,\
                                                                       gallinero2_infestado+gallinero2_no_infestado]})

ax = sns.barplot(x = "Existegallinero", y = "frequency", data = df_bar_plot_pclass_0)
plt.title("Frequency vs Existegallinero (Total Frequency)")
plt.show()
```

```python
#df_bar_plot_pclass = df.drop([each for each in df.columns if not each == "Pclass" or each == "Survived"], axis = 1)
df_bar_plot_pclass = pd.DataFrame({"INFESTADO": ["No", "No", "No", "Yes", "Yes", "Yes"], "Existegallinero": [0, 1, 2, 0, 1, 2],\
                                   "Frequency": [gallinero0_no_infestado, gallinero1_no_infestado, gallinero2_no_infestado,\
                                                gallinero0_infestado, gallinero1_no_infestado, gallinero2_no_infestado]})

ax = sns.barplot(x = "Existegallinero", y = "Frequency", hue = "INFESTADO", data = df_bar_plot_pclass)
plt.title("Frequency vs PClass (Filtered by Survival)")
plt.show()
```
